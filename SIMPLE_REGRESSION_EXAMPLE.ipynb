**Prblm:**“If a student studies more hours, how many marks will they get?”

Feature (X) → Study Hours

Label (y) → Marks

“Since marks are numbers, this is Regression.”

```python
#Import What’s Needed
#pre-built ML tools

from sklearn.linear_model import LinearRegression
import numpy as np
```

```python
#Create Small Data set
x = np.array([1, 2, 3, 4, 5]).reshape(-1, 1)  # study hours : i/t : feature
y = np.array([40, 50, 60, 70, 80])           # marks : label :o/t
```

```python
#Train the Model
model = LinearRegression() #Creates an empty ML model
model.fit(x, y)
```

**What happens here:**

model. fit (x, y)


*  Model looks at study hours (X)
*  Model looks at marks (y)
*  Finds a straight-line relationship
*  Learns the values of m and c in:




          y=mx+c

```python
#Predict for New Value
prediction = model.predict([[10]])
print("Predicted marks:", prediction)
```

```python
#Visual Representation
import matplotlib.pyplot as plt #Used for plotting graphs

plt.scatter(x, y) # Dots represent real data.
plt.plot(x, model.predict(x)) # Draws the regression line
# Labels the graph & Displays it
plt.xlabel("Study Hours")
plt.ylabel("Marks")
plt.show()
```
